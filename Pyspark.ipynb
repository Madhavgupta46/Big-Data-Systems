```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
        .appName("PySpark DataFrame Basics") \
        .getOrCreate()

spark
```

```python
df = spark.read.csv('/content/80cereal.csv', header=True, inferSchema=True)

df.show()
```

```python
df.printSchema()
```

```python
df.describe().show()
```

```python
df.summary().show()
```

```python
df.select("Name").show()
```

```python
df.select("Name", "fat").show()
```

```python
df = df.withColumn("fat_after_increasing_10_ratio", df["fat"] + 10)

df.show()
```

```python
df = df.drop("fat_after_increasing_10_ratio")

df.show()
```

```python
df = df.withColumnRenamed("Name", "Cereal_Name")

df.show()
```
